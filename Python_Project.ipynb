```python
total_spreadsheet = pd.DataFrame()  #all the spreadsheets will be added in total_spreadsheet

for file in list_file:
  if "sales" in file.lower():  #only sales files, lower or upper case
    spreadsheet = pd.read_csv(f'/content/drive/MyDrive/Python/Sales/{file}')
    total_spreadsheet = total_spreadsheet.append(spreadsheet)

 #adds each  excel spreadsheet in one big spreadsheet (total_spreadsheet)

display(total_spreadsheet)
```

```python
total_products = total_spreadsheet.groupby('Product').sum() # group by product column
display(total_products)
```

```python
from numpy import product
import plotly.express as px         # for graphs
```

```python
graph = px.bar(total_products, x = total_products.index, y = 'Items Sold')
graph.show()
```

```python
import matplotlib.pyplot as plt
```

```python
def bar_chart(numbers, labels, pos):
    plt.bar(pos, numbers, color=colors)
    plt.ylabel('Items Sold')
    plt.title('Total Products Sold in Canada')
    plt.xticks(ticks=pos, labels=labels)
    legend = plt.bar(pos, numbers, color=colors)
    plt.legend(legend, labels, loc = 'upper left', title='Product')
    plt.xlabel('Best selling electronics in Canadian stores')
    plt.show()

numbers = [1900, 1645, 1794, 3308, 1725, 1650, 5175]
colors = ['purple', 'brown', 'yellow', 'gray', 'olive', 'blue', 'red']
labels = ['Android', 'Camera', 'Notebook', 'TV', 'SmartWatch', 'Tablet', 'IPhone']
pos = list(range(7))
bar_chart(numbers, labels, pos)
```
